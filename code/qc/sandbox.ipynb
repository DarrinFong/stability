```python
from nilearn.plotting import plot_anat, plot_glass_brain, plot_roi

import nibabel as nib
import numpy as np
import png

import checkerboard
from data2sprite import data2sprite

%matplotlib inline
```

```python
def addNoise(da1):
    tmp = 80*np.random.rand(*da1.shape)
    return da1 + tmp
```

```python
f1 = './transform_Warped.nii.gz'
i1 = nib.load(f1)
d1 = i1.get_data()
```

```python
d2 = addNoise(d1)
i2 = nib.Nifti1Image(d2, i1.affine, i1.header)
```

```python
i3 = checkerboard.checkerNiftii(i1, i2)
d3 = i3.get_data()
```

```python
d4 = data2sprite(d3).astype(np.uint8)
my = png.from_array(d4, 'L').save('tmp.png')
```

```python
plot_anat(i1, cut_coords=[4, -30, -9])
```
